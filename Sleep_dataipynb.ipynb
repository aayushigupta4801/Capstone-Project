**Importing the libraries**

```python
import pandas as pd
import numpy as np
```

```python
minuteSleep_merged = pd.read_csv('minuteSleep_merged.csv')
sleepDay_merged = pd.read_csv('sleepDay_merged.csv')
```

```python
# to convert Date in proper format
minuteSleep_merged['date'] = pd.to_datetime(minuteSleep_merged['date'])
minuteSleep_merged['date'] = pd.to_datetime(minuteSleep_merged['date']).dt.strftime('%d/%m/%Y %H:%M:%S')
minuteSleep_merged['date'] = minuteSleep_merged['date'].str.replace('/', '-')
minuteSleep_merged.head()
```

```python
minuteSleep_merged['date'] = pd.to_datetime(minuteSleep_merged['date'], format='%d-%m-%Y %H:%M:%S')
```

```python
# to convert Date in proper format
sleepDay_merged['SleepDay'] = pd.to_datetime(sleepDay_merged['SleepDay'])
sleepDay_merged['SleepDay'] = pd.to_datetime(sleepDay_merged['SleepDay']).dt.strftime('%d/%m/%Y %H:%M:%S')
sleepDay_merged['SleepDay'] = sleepDay_merged['SleepDay'].str.replace('/', '-')
sleepDay_merged.head()
```

```python
sleepDay_merged['SleepDay'] = pd.to_datetime(sleepDay_merged['SleepDay'], format='%d-%m-%Y %H:%M:%S')
```

```python
sleepDay_merged.info()
```

```python
# Merge the DataFrames on 'Id' and matching dates
Sleep_data = pd.merge(sleepDay_merged, minuteSleep_merged, left_on=['Id', 'SleepDay'], right_on=['Id', 'date'], how='inner')

# Drop the 'date' column if it's no longer needed
Sleep_data = Sleep_data.drop(columns=['date'])
```

```python
Sleep_data.head()
```

**Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
plt.plot(Sleep_data['SleepDay'], Sleep_data['TotalMinutesAsleep'], label='Total Minutes Asleep', color='blue', marker='o')
plt.plot(Sleep_data['SleepDay'], Sleep_data['value'], label='Heart Rate per Second', color='red', marker='x')
plt.title('Total Minutes Asleep and Heart Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Minutes/Heart Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(Sleep_data['TotalMinutesAsleep'], Sleep_data['value'], color='purple')
plt.title('Heart Rate vs. Total Sleep Time')
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Heart Rate per Second')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='TotalTimeInBed', y='value', data=Sleep_data)
plt.title('Box Plot of Heart Rate by Total Time in Bed')
plt.xlabel('Total Time in Bed')
plt.ylabel('Heart Rate per Second')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(Sleep_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Apply Seaborn style
sns.set(style="whitegrid")

# Group data by Intensity and sum the Steps
intensity_steps = Sleep_data.groupby('TotalSleepRecords')['value'].sum()

# Plotting Pie Chart
plt.figure(figsize=(5, 5))
plt.pie(intensity_steps.values, labels=intensity_steps.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Proportion of Total Sleep Records by value Level')
plt.show()
```
