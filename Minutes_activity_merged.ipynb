 **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.express as px

# Group data by Intensity and sum the Steps
intensity_steps = Minutes_activity_merged.groupby('Intensity')['Steps'].sum().reset_index()

# Plotting Pie Chart
fig = px.pie(intensity_steps, values='Steps', names='Intensity', title='Proportion of Total Steps by Intensity Level')
fig.show()
```

```python
# Apply Seaborn style
sns.set(style="whitegrid")

# Group data by Intensity and sum the Steps
intensity_steps = Minutes_activity_merged.groupby('Intensity')['Steps'].sum()

# Plotting Pie Chart
plt.figure(figsize=(5, 5))
plt.pie(intensity_steps, labels=intensity_steps.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Proportion of Total Steps by Intensity Level')
plt.show()
```

```python
# Heatmap using Seaborn
plt.figure(figsize=(6, 6))
corr_matrix = Minutes_activity_merged[['Calories', 'Intensity', 'METs', 'Steps']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Heatmap using Plotly
fig = px.imshow(corr_matrix, text_auto=True, title='Correlation Heatmap')
fig.show()
```
