**Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Histogram using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(Hourly_activity_merged['Calories'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Calories', y='StepTotal', data=Hourly_activity_merged, color='green')
plt.title('Calories vs Steps')
plt.show()
```

```python
# Box plot using Plotly
fig = px.box(Hourly_activity_merged, y='TotalIntensity', title='Box Plot of Intensity')
fig.show()
```

```python

# Heatmap using Seaborn
plt.figure(figsize=(5, 5))
corr_matrix = Hourly_activity_merged[['Calories', 'TotalIntensity', 'AverageIntensity', 'StepTotal']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Bar plot using Plotly
fig = px.bar(Hourly_activity_merged, x=Hourly_activity_merged['ActivityHour'].dt.hour, y='StepTotal', title='Steps by Hour of Day')
fig.show()
```

```python
# Group data by Intensity and sum the Steps
intensity_steps = Hourly_activity_merged.groupby('TotalIntensity')['StepTotal'].sum()

# Plotting Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(intensity_steps, labels=intensity_steps.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Proportion of Total Steps by Intensity Level')
plt.show()
```
